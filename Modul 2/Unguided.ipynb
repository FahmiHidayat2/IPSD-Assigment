4. Drop missing value berdasarkan baris, kolom, imputasi dengan mean, modus, median.

```python
# Drop Missing Value Berdasarkan Baris
drop_baris = df.dropna()
print("\nData setelah menghapus baris dengan missing values:\n", drop_baris)
```

```python
# Drop Missing Value Berdasarkan Kolom
drop_kolom = df.dropna(axis=1)
print("\nData setelah menghapus kolom dengan missing values:\n", drop_kolom)
```

```python
# Imputasi dengan Mean
mean_imputation = df.copy(deep=True)
if 'Time_taken' in mean_imputation.columns:
    mean_imputation['Time_taken'] = mean_imputation['Time_taken'].fillna(mean_imputation['Time_taken'].mean())
print("\nData setelah imputasi dengan mean:\n", mean_imputation)
```

```python
#Imputasi dengan Modus
modus_imputation = df.copy(deep=True)
if 'Time_taken' in modus_imputation.columns:
    modus_imputation['Time_taken'] = modus_imputation['Time_taken'].fillna(modus_imputation['Time_taken'].mode()[0])
print("\nData setelah imputasi dengan modus:\n", modus_imputation)
```

```python
# Imputasi dengan Median
median_imputation = df.copy(deep=True)
if 'Time_taken' in median_imputation.columns:
    median_imputation['Time_taken'] = median_imputation['Time_taken'].fillna(median_imputation['Time_taken'].median())
print("\nData setelah imputasi dengan median:\n", median_imputation)
```
